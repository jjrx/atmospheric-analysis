```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
asthma_df = pd.read_csv('Resources/asthma-ed-visit-rates-lghc-indicator-07.csv')
asthma_df.head(10)
```

```python
filter_out_ca = asthma_df.loc[asthma_df['Geography'] != 'California', :]
filter_out_ca
```

```python
group = filter_out_ca.groupby(['Geography', 'Year'])['Rate'].mean()
```

```python
counties = filter_out_ca['Geography'].unique()
```

```python
rates_df = pd.DataFrame(group).unstack(0)
```

```python
rates_df
```

```python
rates_df.columns
```

```python
plt.plot(rates_df)
plt.xlabel('Year')
plt.ylabel('ED Visit Rates')
```

```python
rates_df.plot(kind='bar')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```

```python
ozone_df = pd.read_csv('Resources/Air_Quality_Measures_on_the_National_Environmental_Health_Tracking_Network.csv')
```

```python
ca_ozone = ozone_df.loc[ozone_df['StateName']=='California',:]
ca_ozone
ca_2011_ozone = ca_ozone.loc[(ca_ozone['ReportYear']>=2011) & (ca_ozone['MeasureName']=='Number of person-days with maximum 8-hour average ozone concentration over the National Ambient Air Quality Standard'),:]
ca_2011_ozone
```

```python
ca_2011_ozone['ReportYear'].max()
```

```python
max_county = ca_2011_ozone['Value'].max()
ca_2011_ozone.loc[ca_2011_ozone['Value']==max_county, 'CountyName']
```

```python
ca_2011_ozone['Value'].value_counts()
```

```python
min_county = ca_2011_ozone['Value'].min()
print(min_county)
ca_2011_ozone.loc[ca_2011_ozone['Value']==min_county, 'CountyName']
```

```python
reduced_df = ca_2011_ozone.loc[(ca_2011_ozone['CountyName']=='San Diego') | (ca_2011_ozone['CountyName']=='Stanislaus') | (ca_2011_ozone['CountyName']=='Alameda') | (ca_2011_ozone['CountyName']=='Los Angeles'), ['ReportYear','CountyName', 'Value']]
reduced_df
```

```python
reduced_group = reduced_df.groupby(['CountyName', 'ReportYear']).mean()
reduced_group.rename(columns={'ReportYear': 'Year'}, inplace=True)
reduced_group
```

```python
reduced_group.plot(kind='bar')
```

```python
plt.plot(reduced_group.unstack(0))
```
